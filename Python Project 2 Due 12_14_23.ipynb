```python
#QUESTION 7: the analysis should include one visualization (or more)
#VISUALIZATION 1

ax = df.groupby(['gender'])['diabetesMed'].count().plot(kind='bar', rot=0, color=['pink', 'lightblue'])
ax.set_xlabel('Gender')
ax.set_ylabel('Count')
ax.set_title('Prescribed Diabetes Medication by Gender')
plt.show()
```

```python
#QUESTION 7: the analysis should include one visualization (or more)
#VISUALIZATION 2

# Create a new DataFrame with the counts of diabetes medication by gender
gender_counts = df.groupby(['gender'])['diabetesMed'].count()

# Create a list of colors
colors = ['pink', 'lightblue']

# Create a pie chart with percentages and colors
plt.pie(gender_counts, labels=gender_counts.index, autopct='%1.1f%%', colors=colors)

# Add a title
plt.title('Prescribed Diabetes Medication by Gender')

# Show the plot
plt.show()
```

```python
#QUESTION 7: the analysis should include one visualization (or more) 
#VISUALIZATION 2-USING SEABORN LIBRARY
# Create a countplot of 'diabetesMed' by 'gender'
#sns.countplot(data=df, x="diabetesMed", hue="gender")
#plt.show()-note original code, I changed colors to pink and blue 
sns.countplot(data=df, x="diabetesMed", hue="gender", palette=["pink", "lightblue"])
plt.show()
```
